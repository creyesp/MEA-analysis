```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
file_path = '../../reports/spreadsheets/MR-0092t2_allnd_valid.csv'
df_valid = pd.read_csv(file_path)
df_valid.columns.to_list()
```

```python
g = sns.catplot(
    x='nd_filter',
    y='RF_PeakFrame',
    hue='RF_STAtype',
    col='FL_255_type',
    data=df_valid,
)
g.set(ylim=(-0.15, -0.040))
```

```python
f, ax = plt.subplots(figsize=(10, 10 ))
for kname, kdf in df_valid.groupby('nd_filter'):
    filter_valid = kdf['RF_valid']
    kwargs = {'bins':40, 'range':(0,40), 'density':True, 'label': kname}
    ax.hist(kdf['RF_snr'][~filter_valid], histtype='step', **kwargs)
    ax.hist(kdf['RF_snr'][filter_valid], alpha=0.5, histtype='stepfilled', **kwargs)
ax.legend()
```
